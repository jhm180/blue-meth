```python
###################
##### GENERATES A CSV OF CURRENT MARKET PRICES THAT IS INPUT INTO THE EXCEL-BASED DVT TOOL 
###################

df['TotalPrice'] = df['PricePerCarat'] * df['Carat'] #l.Price represents price per carat

#creates set of column headers to be used to combine 
clars_a = ['IF_#', 'VVS1_#', 'VVS2_#', 'VS1_#', 'VS2_#', 'SI1_#', 'SI2_#', 'SI3_#', 'I1_#', 'I2_#','I3_#']
clars_b = ['BLANK_COL', 'IF_#', 'VVS1_#', 'VVS2_#', 'VS1_#', 'VS2_#', 'SI1_#', 'SI2_#', 'S3_#', 'I1_#', 'I2_#','I3_#']
clars_tot = clars + clars_b
df_clars_tot = pd.DataFrame(clars_tot)

#clear csv file from previous data dumps
f = open("/home/oliver/Dropbox/whitepine/DVT_output.csv", "w")
f.truncate()
f.close()

num = 10  #Defines lowest N of a category to take the average of. EG num = 10 means the lowest 10 prices will be averaged
toss = 2  #Defines number of prices to be thrown out before averaging lowest N. EG toss = 2 means the lowest 2 prices will be ignored when average of lowest N is calculated. 

#writer = pd.ExcelWriter('/home/oliver/Dropbox/whitepine/DVT_output.xlsx')

def DVT_Output(min_carat,max_carat):
    df_number_of_stones = pd.DataFrame(0, index=colors, columns=clars)  
    df_avg_of_lowest_pct_back = pd.DataFrame(0, index=colors, columns=clars_tot, dtype=np.float64) 
    
    def df_sub_maker(color,clar,min_carat,max_carat): 
        return df[(df['Color'] == color ) & (df['Clarity'] == clar) & (df['Carat'] >= min_carat) & (df['Carat'] <= max_carat) \
        & (df['Fluor'].isin(fluor_none_or_faint))]
    
    def find_number_of_stones(color,clar,min_carat,max_carat):
        return len(df_subset)

    def average_of_lowest_pct_back(color,clar,min_carat,max_carat):
        df_subset_sorted = df_subset.sort('PctRap')
        endidx = min(len(df_subset_sorted),num) 
        startidx = min(len(df_subset_sorted),toss)
        if len(df_subset_sorted) == 0:
            return 0
        elif len(df_subset_sorted) <= num:
            return "%.1f" % (df_subset_sorted['PctRap'].mean()*100)   
        else:
            return "%.1f" % (df_subset_sorted['PctRap'].iloc[toss:endidx].mean()*100)    
    
    for i in colors:
        color = i
        for l in clars: 
            clar = l
            df_subset = df_sub_maker(color,clar,min_carat,max_carat)
            df_number_of_stones.ix[color,clar] = find_number_of_stones(color,clar,min_carat,max_carat)
            df_avg_of_lowest_pct_back.ix[color,clar] = average_of_lowest_pct_back(color,clar,min_carat,max_carat)
    
    df_number_of_stones.columns = clars_a #renames column headers in df_number_of_stones to append 
    
    for i in clars_a: #appends data in df_number_of_stones to avg_of_lowest_pct_back, making data eaiser to dump into excel
        clar = i
        df_avg_of_lowest_pct_back[clar] =  df_number_of_stones[clar]
    
    df_table_header = pd.DataFrame(["", "size bins: minimum carat weight = %s and maximum carat weight = %s" % (str(min_carat),str(max_carat))])
    df_table_header.to_csv('/home/oliver/Dropbox/whitepine/DVT_output.csv', mode='a', header=False) #writes table header to csv file
    df_clars_tot.T.to_csv('/home/oliver/Dropbox/whitepine/DVT_output.csv', mode='a', header=False) #writes column headers for tables 
    df_avg_of_lowest_pct_back.to_csv('/home/oliver/Dropbox/whitepine/DVT_output.csv', mode='a', header=False) #writes table for each size band determined by min carat, max carat


#min/max carat weights shown below are the WP 'core price bands' defined by Ben
DVT_Output(0.30, 0.35)        
DVT_Output(0.40, 0.45)
DVT_Output(0.50, 0.59)
DVT_Output(0.74, 0.79)
DVT_Output(0.90, 0.95)
DVT_Output(1.00, 1.09)    
DVT_Output(1.50, 1.59)
DVT_Output(2.00, 2.19)


#Note: Can use the code below to output an excel file with data table on every tab - the above method is fastier to copy and paste into DVT pricing tool
#writer = ExcelWriter('path_to_file.xlsx')
    #excel_tab_name = "%s to %s cts" %(str(min_carat),str(max_carat))
    #df_avg_of_lowest_pct_back.to_excel(writer, sheet_name=excel_tab_name)
#writer.save()

```
