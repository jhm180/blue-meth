```python
import pandas as pd

import MySQLdb

c = MySQLdb.connect('localhost', 'root', '3lihu_r007', 'rapnet_listings')

import pandas.io.sql as psql

df = psql.read_frame("""select l.LotNum, convert(l.Price, decimal(10,4)) as Price, l.Shape, convert(l.Carat, decimal(10,6)) as Carat, l.Color, l.Clarity, l.CutGrade, l.PctRap, l.Cert, l.Depth, l.TableWidth, l.Girdle, 
l.Culet, l.Polish, l.Sym, l.Fluor, l.Meas, l.RapnetComment, l.NumStones, l.CertNum, l.StockNum, l.Make, l.Date, l.City,
l.State, l.Country, l.Image from active_listing l where l.Cert like 'GIA' and l.Country like 'USA' 
and not l.CutGrade is null and l.CutGrade <> ''
and not l.Polish is null and l.Polish <> ''
and not l.Sym is null and l.Sym <> ''
and not l.Fluor is null and l.Fluor <> ''
and not l.Price is null and l.Price <> ''""", c, index_col = 'LotNum')
```

```python
filtered_rows = set()
rank_dict = {'Excellent':0, 'Very Good':1, 'Good':2}

ct = 0
for idx, row in df.iterrows():
    ct += 1
    cut = rank_dict.get(row['CutGrade'], -1)
    polish = rank_dict.get(row['Polish'], -1)
    symmetry = rank_dict.get(row['Sym'], -1)
    
    if cut >= 0 and polish >= 0 and symmetry >= 0 \
    and abs(cut - polish) <= 1 and abs(polish - symmetry) <= 1 and abs(cut - symmetry) <= 1:
        filtered_rows.add(idx)
df2 = df[df.index.map(lambda x: x in filtered_rows)]
```

```python
from IPython.core.display import HTML
def dump_bucket_info(minCarat, maxCarat):
    df_subset = df2[(df2.Carat >= minCarat) & (df2.Carat < maxCarat)]
    if len(df_subset) > 0:
        def dump_lowest_n(n, grp):
            firstidx = 0
            sgrp = grp.sort('Price')
            if len(sgrp) > n and (sgrp.iloc[0]['Price'] / sgrp.iloc[1]['Price']) < 0.95:
                firstidx = 1 # throw out first result if more than 5% less than next one
            maxct = min(firstidx+n, len(sgrp))
            v = sgrp.iloc[firstidx:maxct][['Price']]
            return pd.DataFrame({'avg price':v.mean(), 'group count': len(sgrp)})
        grouped = df_subset.groupby(['CutGrade', 'Polish', 'Sym', 'Fluor'])#.aggregate({'Price': [np.mean, len]})
        for grpSize in [5,10]:
            dfGrouped = grouped.apply(lambda grp: dump_lowest_n(grpSize, grp))
            display(HTML("<h4>Average of lowest {0} prices for {1} to {2} carats</h4>".format(grpSize, minCarat, maxCarat)))
            display(HTML(dfGrouped.to_html()))
    else:
        print 'No results for range {0} to {1} carats'.format(minCarat, maxCarat)
        
dump_bucket_info(0.5, 0.7)
dump_bucket_info(0.7, 0.9)
dump_bucket_info(0.9, 1.0)
dump_bucket_info(1.0, 1.5)
dump_bucket_info(1.5, 2.0)
dump_bucket_info(2.0, 3.0)
```
