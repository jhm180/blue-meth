```python
colors = ['D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M']
clars = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1', 'I2']

num = 10  #Defines lowest N of a category to take the average of. EG num = 10 means the lowest 10 prices will be averaged
toss = 2  #Defines number of prices to be thrown out before averaging lowest N. EG toss = 2 means the lowest 2 prices will be ignored when average of lowest N is calculated. 

def summary_statistics(min_carat,max_carat):
    df_number_of_stones = pd.DataFrame(0, index=colors, columns=clars)  
    df_med_price = pd.DataFrame(0, index=colors, columns=clars)  
    df_avg_price = pd.DataFrame(0, index=colors, columns=clars)  
    df_avg_pct_back = pd.DataFrame(0, index=colors, columns=clars, dtype=np.float64)  
    df_avg_of_lowest = pd.DataFrame(0, index=colors, columns=clars) 
    df_avg_of_lowest_pct_back = pd.DataFrame(0, index=colors, columns=clars, dtype=np.float64) 
    
    def df_sub_maker(color,clar,min_carat,max_carat):
         return df[(df['Color'] == color ) & (df['Clarity'] == clar) & (df['Carat'] >= min_carat) & (df['Carat'] <= max_carat)]
    
    def find_number_of_stones(color,clar,min_carat,max_carat):
        return len(df_subset)
    
    def average(color,clar,min_carat,max_carat):
        if len(df_subset) == 0:
            return 0
        else: 
            return df_subset['Price'].mean()
    
    def avg_pct_back(color,clar,min_carat,max_carat):
        if len(df_subset) == 0:
            return 0
        else: 
            return "%.1f" %(df_subset['PctRap'].mean()*100)
    
    def median(color,clar,min_carat,max_carat):
        if len(df_subset) == 0:
            return 0
        else: 
            return df_subset['Price'].median()
        
    def average_of_lowest(color,clar,min_carat,max_carat):
        df_subset_sorted = df_subset.sort('Price')
        endidx = min(len(df_subset_sorted),num) 
        startidx = min(len(df_subset_sorted),toss)
        if len(df_subset_sorted) == 0:
            return 0
        elif len(df_subset_sorted) <= num:
            return df_subset_sorted['Price'].mean()
        else:
            return df_subset_sorted['Price'].iloc[toss:endidx].mean()   

    def average_of_lowest_pct_back(color,clar,min_carat,max_carat):
        df_subset_sorted = df_subset.sort('PctRap')
        endidx = min(len(df_subset_sorted),num) 
        startidx = min(len(df_subset_sorted),toss)
        if len(df_subset_sorted) == 0:
            return 0
        elif len(df_subset_sorted) <= num:
            return "%.1f" % (df_subset_sorted['PctRap'].mean()*100)   
        else:
            return "%.1f" % (df_subset_sorted['PctRap'].iloc[toss:endidx].mean()*100)    

    for i in colors:
        color = i
        for l in clars: 
            clar = l
            df_subset = df_sub_maker(color,clar,min_carat,max_carat)
            df_number_of_stones.ix[color,clar] = find_number_of_stones(color,clar,min_carat,max_carat)
            df_avg_price.ix[color,clar] = average(color,clar,min_carat,max_carat)
            df_med_price.ix[color,clar] = median(color,clar,min_carat,max_carat)
            df_avg_of_lowest.ix[color,clar] = average_of_lowest(color,clar,min_carat,max_carat)
            df_avg_pct_back.ix[color,clar] = avg_pct_back(color,clar,min_carat,max_carat)
            df_avg_of_lowest_pct_back.ix[color,clar] = average_of_lowest_pct_back(color,clar,min_carat,max_carat)
            
    print "size bins: minimum carat weight = %s and maximum carat weight = %s  \n" % (min_carat,max_carat)
    print "\n Number of stones \n"
    print df_number_of_stones
    print "\n Average price of lowest %s stones, excluding the lowest %s \n" % (num,toss)
    print df_avg_of_lowest
    print "\n Average pct back of lowest %s stones, excluding the lowest %s \n" % (num,toss)
    print df_avg_of_lowest_pct_back
    print "\n Average price of all stones in category\n"
    print df_avg_price 
    print "\n Average pct back of all stones in category \n"
    print df_avg_pct_back
    print "\n Median price of all stones in category\n"
    print df_med_price 
    

summary_statistics(1.0, 1.1)    
#summary_statistics(1.1, 1.2)
#summary_statistics(1.2, 1.3)
#summary_statistics(1.3, 1.4)
#summary_statistics(1.4, 1.5)
```
